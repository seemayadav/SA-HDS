```python
def get_summary(con, query):
    prompt = f"""
You are a helpful assistant that generates **concise, fluent, and informative abstractive summaries** from extractive summaries and user queries.

### Example
Query: What are the current updates from the development team?

Dialogue:
Project Manager: Good morning everyone. Let's start with the status updates. Sarah, how's the frontend module coming along?
Sarah: Morning! The UI design is finalized, and I’ve completed about 80% of the implementation. Just some responsive layout issues left.
Project Manager: Great. Make sure to address those before the review. Tom, any updates on the backend?
Tom: Yes, the API integration is done. I’m currently working on optimizing the database queries. Should be completed by tomorrow.
Sarah: Tom, do you need any changes from my side for API endpoints?
Tom: No, everything looks good for now. I’ll let you know if anything comes up.
Project Manager: Excellent. Once both parts are stable, we’ll move to system testing. Let’s aim to start that by Friday.
Tester (Priya): That works. I’ll prepare the test cases by Thursday.
Project Manager: Perfect. That’s all for today. Let’s catch up in the next stand-up.

Abstractive Summary:
The team discussed progress on the frontend and backend modules. Sarah is nearly done with the UI, while Tom has completed API work and is refining database performance. Testing is scheduled to begin Friday, with test cases being prepared in advance.

### Now your turn
Query: {query}

Extractive Summary:
{con}

Instruction: Based on the extractive summary and query above, generate a concise, fluent, and informative abstractive summary."""

    chat_completion = client.chat.completions.create(
        model="llama3-70b-8192",
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant that summarizes dialogues concisely."
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=1,
        max_tokens=1024,
        top_p=0.8,
        stop=None,
        stream=False,
    )
    return chat_completion.choices[0].message.content
```

```python

nltk.download('punkt')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
model = SentenceTransformer('all-MiniLM-L6-v2')
# Preprocessing

import re
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')


def clean_text(text: str) -> str:
    text = re.sub(r'\{.*?\}', '', text)  # remove ASR markers like {vocalsound}, {disfmarker}
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra whitespace
    return text


# 1. Preprocessing function
def clean_dialogue_text(dialogue):
    dialogue = re.sub(r'\{.*?\}', '', dialogue)  # remove {vocalsound}, {disfmarker}, etc.
    dialogue = re.sub(r'[^\w\s:.]', '', dialogue)  # remove extra symbols
    dialogue = re.sub(r'\s+', ' ', dialogue)  # collapse multiple spaces
    lines = dialogue.split('\n')

    #print("Dialogues :", lines)

    cleaned_lines = []
    for line in lines:
        if ':' in line:
            line = line.split(':', 1)[1]  # remove speaker
        tokens = word_tokenize(line.lower())
        tokens = [w for w in tokens if w.isalpha() and w not in stop_words]
        tagged = pos_tag(tokens)
        # Keep only nouns and adjectives
        filtered = [word for word, tag in tagged if tag.startswith('N') or tag.startswith('J')]
        cleaned_lines.append(" ".join(filtered))
    return cleaned_lines

# 2. Keyword extraction

# Standard stopwords
standard_stopwords = set(stopwords.words('english'))

# Add custom filler/disfluency tokens
disfluency_tokens = {
    "uh", "um", "uhm", "erm", "mm", "hmm", "ah", "oh", "eh", "ha", "huh", "uhuh", "mhm", "uh-uh",
    "yeah", "yes", "no", "okay", "ok", "alright", "right", "sure", "maybe", "well", "so", "like",
    "just", "really", "basically", "actually", "literally", "kinda", "sorta", "somewhat", "guess",
    "you", "we", "they", "i", "he", "she", "it", "me", "him", "her", "them", "us",
    "know", "see", "let", "say", "get", "go", "come", "do", "did", "does", "done", "doing",
    "be", "am", "is", "are", "was", "were", "been", "being",
    "have", "has", "had", "having",
    "will", "would", "can", "could", "shall", "should", "may", "might", "must",
    "then", "there", "here", "that", "this", "those", "these",
    "more", "much", "very", "true", "not", "no", "yeah", "ok", "okay", "alright",
    "uh-huh", "uh-oh", "mmm", "uhuh", "huh", "um-hum", "uhuh", "ain't", "gonna", "wanna", "gotta",
    "thing", "stuff", "something", "anything", "everything", "nothing", "thingy", "things",
    "gon", "gonna", "wanna"
}
custom_stopwords = standard_stopwords.union(disfluency_tokens)

def clean_and_extract_filtered_words(text):
    text = re.sub(r'\{.*?\}', '', text)  # remove tags like {vocalsound}
    text = re.sub(r'\b[A-Z][A-Za-z ]{1,30}:\s*', '', text, flags=re.MULTILINE)  # remove speaker names

    sentences = nltk.sent_tokenize(text)
    content_sentences = []

    for sent in sentences:
        words = nltk.word_tokenize(sent)
        pos_tags = nltk.pos_tag(words)

        filtered_words = [
            word.lower()
            for word, tag in pos_tags
            if tag.startswith(('NN', 'VB', 'JJ', 'RB')) and
               word.lower() not in custom_stopwords and
               word.isalpha()
        ]

        if filtered_words:
            content_sentences.append(" ".join(filtered_words))

    return content_sentences

def extract_keywords_filtered(dialogue_texts, top_n=50):
    all_keywords = []

    for dialogue in dialogue_texts:
        lines = dialogue.strip().split('\n')
        if not lines:
            continue

        query = lines[0].strip()
        content = "\n".join(lines[1:]).strip()

        cleaned_sentences = clean_and_extract_filtered_words(content)
        if not cleaned_sentences:
            continue

        vectorizer = TfidfVectorizer(max_features=1000)
        tfidf_matrix = vectorizer.fit_transform(cleaned_sentences)
        terms = vectorizer.get_feature_names_out()
        scores = tfidf_matrix.mean(axis=0).A1

        top_indices = scores.argsort()[::-1]
        top_keywords = [terms[i] for i in top_indices[:top_n]]

        # Rerank based on similarity
        if query:
            query_emb = model.encode([query])
            keyword_embs = model.encode(top_keywords)
            sims = cosine_similarity(query_emb, keyword_embs)[0]
            top_keywords = [kw for _, kw in sorted(zip(sims, top_keywords), reverse=True)]

        all_keywords.append({
            "query": query,
            "top_keywords": top_keywords
        })

    return all_keywords
```

```python

def preprocess_transcript(text: str) -> Tuple[List[str], List[str]]:
    pattern = r'([A-Za-z\s]+):\s*(.*?)(?=(?:[A-Za-z\s]+:)|\Z)'  # Match speaker + content
    matches = re.findall(pattern, text, re.DOTALL)

    structured_sentences = []
    speakers = []

    for speaker, content in matches:
        speaker = speaker.strip()
        content = clean_text(content.strip())  # 🔁 Clean text here
        if not content:
            continue
        sentences = nltk.sent_tokenize(content)  # Sentence tokenization after cleaning
        structured_sentences.extend(sentences)
        speakers.extend([speaker] * len(sentences))

    return structured_sentences, speakers


def compute_embeddings(sentences: List[str]):
    return model.encode(sentences, convert_to_tensor=False)


def compute_informativeness(embeddings, selected_idx):
    return np.sum([np.linalg.norm(embeddings[i]) for i in selected_idx])


def compute_query_coverage(embeddings, query_embed, selected_idx):
    return np.mean([cosine_similarity([embeddings[i]], [query_embed])[0][0] for i in selected_idx])


def compute_weighted_informativeness_query(embeddings, query_embed, selected_idx):
    summary_embeds = [embeddings[i] for i in selected_idx]
    informativeness = np.sum([np.linalg.norm(e) for e in summary_embeds])
    query_rel = np.mean([cosine_similarity([e], [query_embed])[0][0] for e in summary_embeds])
    return 0.6 * informativeness + 0.4 * query_rel

def compute_speaker_diversity(speakers, selected_idx):
    return len(set(speakers[i] for i in selected_idx)) / len(selected_idx)

def compute_redundancy(embeddings, selected_idx):
    if len(selected_idx) < 2:
        return 0.0
    selected = [embeddings[i] for i in selected_idx]
    sim_matrix = cosine_similarity(selected)
    np.fill_diagonal(sim_matrix, 0)
    return np.sum(sim_matrix) / (len(selected_idx) * (len(selected_idx) - 1))

def compute_length_penalty(selected_idx, max_len):
    return abs(len(selected_idx) - max_len)



```

```python
class HoneyBadger:
    def __init__(self, sentences: List[str], speakers: List[str], query: str, pop_size: int, max_iter: int, max_len: int):
        self.sentences = sentences
        self.speakers = speakers
        self.query = query
        self.N = pop_size
        self.T = max_iter
        self.dim = len(sentences)
        self.max_len = max_len
        self.population = [self.init_solution() for _ in range(self.N)]
        self.archive = []
        self.embeddings = compute_embeddings(sentences)
        self.query_embed = model.encode(query)

    def init_solution(self):
        return sorted(random.sample(range(self.dim), random.randint(1, self.max_len)))

    def evaluate(self, sol):
        inf = compute_informativeness(self.embeddings, sol)
        qc = compute_query_coverage(self.embeddings, self.query_embed, sol)
        div = compute_speaker_diversity(self.speakers, sol)
        red = compute_redundancy(self.embeddings, sol)
        lp = compute_length_penalty(sol, self.max_len)
        group1 = 0.4 * inf + 0.3 * qc + 0.3 * div
        group2 = 0.5 * red + 0.5 * lp
        return (group1, -group2)

    def dominates(self, a, b):
        return all(x >= y for x, y in zip(a, b)) and any(x > y for x, y in zip(a, b))

    def update_archive(self, candidates):
        for sol, fit in candidates:
            dominated = False
            self.archive = [arc for arc in self.archive if not self.dominates(fit, arc[1])]
            for _, arc_fit in self.archive:
                if self.dominates(arc_fit, fit):
                    dominated = True
                    break
            if not dominated:
                self.archive.append((sol, fit))

    def optimize(self):
        alpha = 0.8
        beta = 1.5
        for t in range(self.T):
            candidates = []
            for i in range(self.N):
                current = self.population[i]
                leader = random.choice(self.archive)[0] if self.archive else random.choice(self.population)
                dist = abs(len(current) - len(leader))
                I = (alpha * np.exp(-random.random() * beta)) / (dist + 1e-6)

                # --- Exploration vs Exploitation ---
                if random.random() < 0.5:
                    # 🐾 Digging behavior: Local search (remove or add one sentence)
                    new_sol = current[:]
                    if len(new_sol) > 1 and random.random() < 0.5:
                        new_sol.remove(random.choice(new_sol))
                    elif len(new_sol) < self.max_len:
                        new_sent = random.choice(list(set(range(self.dim)) - set(new_sol)))
                        new_sol.append(new_sent)
                else:
                    # 🍯 Honey-searching behavior: Move towards leader (combine current and leader)
                    new_sol = list(set(current + leader))
                    if len(new_sol) > self.max_len:
                        new_sol = random.sample(new_sol, self.max_len)

                new_sol = sorted(set(new_sol))
                fit = self.evaluate(new_sol)
                candidates.append((new_sol, fit))

            self.update_archive(candidates)
            alpha *= 0.98

        return self.archive
```

```python
# Evaluation with ROUGE
from rouge_score import rouge_scorer

def compute_rouge_scores(summary: str, reference: str) -> dict:
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    return scorer.score(reference, summary)

# Visualization of Pareto Front

def plot_pareto_front(archive):
    x = [-f[1] for (_, f) in archive]  # -group2 = penalty
    y = [f[0] for (_, f) in archive]   # group1 = quality
    plt.figure(figsize=(6, 4))
    plt.scatter(x, y, c='blue')
    plt.xlabel('Penalty Score (Redundancy + Length)')
    plt.ylabel('Summary Quality (Inf + Div + Query)')
    plt.title('Pareto Front of Summary Candidates')
    plt.grid(True)
    plt.tight_layout()
    plt.show()



def separate_query_and_dialogue(text):
    lines = text.strip().split("\n")
    query = lines[0].strip()
    dialogue = "\n".join(lines[1:]).strip()
    return query, dialogue

# Semantic filtering helper using CPU-safe cosine similarity
def semantic_score(summary, query_embed):
    emb = model.encode(summary)
    return cosine_similarity([emb], [query_embed])[0][0]
```

```python
# Initialize

RUNS = 5
INSTANCES = 10   # number of test examples per run

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
smoothie = SmoothingFunction().method4

extractive_scores = {'rouge1': [], 'rouge2': [], 'rougeL': [], 'bleu': [], 'bert': []}
abstractive_scores = {'rouge1': [], 'rouge2': [], 'rougeL': [], 'bleu': [], 'bert': []}
results = []

for i in range(0, 5):  # loop through few samples for testing
    text = train_df['input'][i]
    gold_summary = train_df['output'][i]
    query, raw_text = separate_query_and_dialogue(text)
    sentences, speakers = preprocess_transcript(raw_text)

    # Get CPU-safe query embedding
    query_emb = model.encode(query)

    # Filter relevant sentences
    filtered_sentences = []
    filtered_speakers = []
    for sent, spk in zip(sentences, speakers):
        sent_emb = model.encode(sent)
        sim = cosine_similarity([sent_emb], [query_emb])[0][0]
        # find the keywords matching the query
        matching_entry = next((item for item in domain_keywords if item['query'] == query), None)
        if matching_entry:
          domain_keywords_list = matching_entry['top_keywords']
        else:
          domain_keywords_list = []

        # use in your loop
        keyword_match = any(kw.lower() in sent.lower() for kw in domain_keywords_list)
        #keyword_match = any(kw.lower() in sent.lower() for kw in domain_keywords)

        if sim >= 0.25 or keyword_match:
          filtered_sentences.append(sent)
          filtered_speakers.append(spk)

    print("Filtered sentences are :", filtered_sentences)

    # Run HBA optimization
    hba = HoneyBadger(filtered_sentences, filtered_speakers, query, pop_size= 20, max_iter=20, max_len=20)

    pareto_front = hba.optimize()

    # Construct Pareto candidates
    candidates = [
        (' '.join([sentences[i] for i in idx]), fit)
        for idx, fit in pareto_front
    ]

    # Pick best semantic match from candidates
    extractive_summary = max(candidates, key=lambda x: semantic_score(x[0], hba.query_embed))[0]

    # Scores for Extractive
    e_rouge = scorer.score(gold_summary, extractive_summary)
    e_bleu = sentence_bleu([nltk.word_tokenize(gold_summary)],
                           nltk.word_tokenize(extractive_summary),
                           smoothing_function=smoothie)
    _, _, e_bert = bert_score([extractive_summary], [gold_summary], lang='en', rescale_with_baseline=True)

    # Generate Abstractive summary
    abstractive_summary = get_summary(extractive_summary, query)

    # Scores for Abstractive
    a_rouge = scorer.score(gold_summary, abstractive_summary)
    a_bleu = sentence_bleu([nltk.word_tokenize(gold_summary)],
                           nltk.word_tokenize(abstractive_summary),
                           smoothing_function=smoothie)
    _, _, a_bert = bert_score([abstractive_summary], [gold_summary], lang='en', rescale_with_baseline=True)

    # Store metrics
    extractive_scores['rouge1'].append(e_rouge['rouge1'].fmeasure)
    extractive_scores['rouge2'].append(e_rouge['rouge2'].fmeasure)
    extractive_scores['rougeL'].append(e_rouge['rougeL'].fmeasure)
    extractive_scores['bleu'].append(e_bleu)
    extractive_scores['bert'].append(e_bert.item())

    abstractive_scores['rouge1'].append(a_rouge['rouge1'].fmeasure)
    abstractive_scores['rouge2'].append(a_rouge['rouge2'].fmeasure)
    abstractive_scores['rougeL'].append(a_rouge['rougeL'].fmeasure)
    abstractive_scores['bleu'].append(a_bleu)
    abstractive_scores['bert'].append(a_bert.item())

    results.append({
        'dialogue_id': i,
        'query': query,
        'gold_summary': gold_summary,
        'extractive_summary': extractive_summary,
        'abstractive_summary': abstractive_summary,
        'extractive_metrics': {
            'rouge1': e_rouge['rouge1'].fmeasure,
            'rouge2': e_rouge['rouge2'].fmeasure,
            'rougeL': e_rouge['rougeL'].fmeasure,
            'bleu': e_bleu,
            'bert': e_bert.item()
        },
        'abstractive_metrics': {
            'rouge1': a_rouge['rouge1'].fmeasure,
            'rouge2': a_rouge['rouge2'].fmeasure,
            'rougeL': a_rouge['rougeL'].fmeasure,
            'bleu': a_bleu,
            'bert': a_bert.item()
        }
    })

# Compute Averages
def average(lst):
    return sum(lst) / len(lst) if lst else 0.0

print("\n=== Average Extractive Scores ===")
for metric in extractive_scores:
    print(f"{metric.upper()}: {average(extractive_scores[metric]):.4f}")

print("\n=== Average Abstractive Scores ===")
for metric in abstractive_scores:
    print(f"{metric.upper()}: {average(abstractive_scores[metric]):.4f}")

with open("summary_eval_results.json", "w") as f:
    json.dump(results, f, indent=2)
```
